# Derive new colums

```python
data['Discount']= data['List Price']*data['Discount Percent']/100
data['Discount']
```

```python
data.columns
```

```python
# Selling Price
# Selling price = List Price - Discount
data['Selling_Price'] = data['List Price']-data['Discount']
```

```python
data.columns
```

```python
# Profit = Sellng price - Cost Price
data['Profit']= data['Selling_Price']- data['cost price']
```

```python
# we will convert the Order date from object to datetime
data['Order Date']= pd.to_datetime(data['Order Date'],format="%Y-%m-%d")
```
